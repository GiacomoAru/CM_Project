```python
from plot_utils import *
from math_utils import *
from pathlib import Path

seed = 456
random.seed(seed)
```

```python
methods = [
    'rand_u',           # Uniform Distribution
    'rand_n',           # Normal Distribution
    'scaled_u',         # Scaled Uniform Distribution
    'scaled_n',         # Scaled Normal Distribution
    'sketching_g',      # Sketching with Gaussian Distribution
    'sketching_b',      # Sketching with Bernoulli Distribution
    'semi-orthogonal'   # Semi-Orthogonal via Householder
]
epsilon = [
    1e-08,
    10e-16
]
ks = [
    1,
    2,
    5,
    10
]
```

```python
for num in range(10):
    mnist_dir = Path(f'./data/img/mnist/{num}')

    imgs = random.sample([f for f in mnist_dir.iterdir()], 2)

    for img in imgs:
        A = load_image_as_grayscale_matrix(img)
        for meth in methods:
            for k in ks:
                for eps in epsilon:
                    start(A, k, 'mnist', str(num) + "_" + img.name,  f'{k}_{meth}_{eps}', meth, epsilon=eps, seed=seed)
```
